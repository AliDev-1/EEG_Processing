```python
# Import necessary libraries
from scipy.io import loadmat, savemat  # Used to load and save MATLAB (.mat) files.
import os  # Provides tools to interact with the file system.
import numpy as np  # Library for numerical operations, especially on arrays.

# Specify the directory containing the input .mat files
data_dir = 'C:/Users/DELL/mBTI_Data'  
# This is the folder where your raw EEG .mat files are stored. Update it to your actual path.

# Specify the directory to save the cleaned .mat files
output_dir = 'C:/Users/DELL/mBTI_Data/Data_Cleaned'  
# Update this to the folder where you want to save the cleaned data.

# Ensure the output directory exists; if not, create it
os.makedirs(output_dir, exist_ok=True)  
# `os.makedirs` creates the directory and its parent directories if they don't exist.
# `exist_ok=True` avoids errors if the directory already exists.

# Loop through each file in the data directory
for file in os.listdir(data_dir):  # `os.listdir` lists all files and folders in the directory.
    if file.endswith('.mat'):  # Process only files with a .mat extension.
        # Full path to the current .mat file
        file_path = os.path.join(data_dir, file)  
        
        # Load the contents of the .mat file
        data = loadmat(file_path, struct_as_record=False, squeeze_me=True)  
        # `struct_as_record=False` loads MATLAB structs as objects.
        # `squeeze_me=True` simplifies arrays, removing unnecessary dimensions.

        # Check if the file contains an 'EEG' structure
        if 'EEG' in data:  
            eeg = data['EEG']  # Extract the 'EEG' structure from the .mat file.
            print(f"Processing file: {file}")  # Indicate which file is being processed.

            # Check if the 'EEG' structure has the required attributes
            if hasattr(eeg, 'data') and hasattr(eeg, 'icaweights') and hasattr(eeg, 'icasphere') and hasattr(eeg, 'bad_ICAs'):
                signal = eeg.data  # EEG signal (channels x time points x trials).
                ica_weights = eeg.icaweights  # ICA weights for separating components.
                ica_sphere = eeg.icasphere  # ICA sphere matrix for data normalization.
                bad_ICs = eeg.bad_ICAs  # List of bad Independent Components (ICs) to remove.

                # Skip files with missing data
                if signal is None or ica_weights is None or ica_sphere is None or bad_ICs is None:
                    print(f"Missing data in {file}. Skipping.")
                    continue

                # Ensure 'bad_ICs' is a list of integers
                try:
                    if isinstance(bad_ICs, (list, np.ndarray)):
                        # Convert valid elements to integers
                        bad_ICs = [int(ic) for ic in bad_ICs if isinstance(ic, (int, float, np.integer))]
                    elif isinstance(bad_ICs, (int, float, np.integer)):
                        bad_ICs = [int(bad_ICs)]  # If it's a single value, convert to a one-element list.
                    else:
                        # If the format of 'bad_ICs' is unexpected, skip processing
                        print(f"Unexpected format for bad_ICs in {file}. Skipping IC removal.")
                        bad_ICs = []
                except Exception as e:
                    # Handle errors during conversion of 'bad_ICs'
                    print(f"Error processing bad_ICs in {file}: {e}")
                    bad_ICs = []

                # Remove contributions of bad ICs if any are specified
                if bad_ICs:
                    # Calculate ICA inverse weights
                    icawinv = np.dot(np.linalg.pinv(ica_weights), ica_sphere)  

                    # Loop through each trial to process the data
                    for trial in range(signal.shape[2]):  # Iterate over trials.
                        for ic in bad_ICs:  # Iterate over the bad ICs.
                            # Subtract the contribution of the bad IC from the signal
                            signal[:, :, trial] -= np.outer(
                                icawinv[:, ic], np.dot(ica_weights[ic, :], signal[:, :, trial])
                            )

                # Save the cleaned data to the output directory
                output_path = os.path.join(output_dir, f"cleaned_{file}")  
                # Construct the file path for the cleaned data.

                savemat(output_path, {'cleaned_data': signal})  
                # Save the cleaned EEG signal to a new .mat file.
                print(f"Cleaned data saved: {output_path}")  # Indicate successful saving.

            else:
                # If required attributes are missing in the 'EEG' structure, skip the file
                print(f"Required fields missing in {file}. Skipping.")
        else:
            # If no 'EEG' structure is found in the file, skip it
            print(f"No EEG structure found in {file}. Skipping.")
```

```python
import os  # Provides tools to work with the operating system, such as file paths and directories.
import numpy as np  # Library for handling numerical operations and arrays.
from scipy.io import loadmat  # Function to load MATLAB (.mat) files into Python.

# Specify the directory containing the cleaned .mat files
output_dir = 'C:/Users/DELL/mBTI_Data/Data_Cleaned'  
# This is the folder where the cleaned EEG data files are stored.

# Initialize lists to store the data and subject IDs
all_signals = []  # This list will store all the trials from all files.
subject_ids = []  # This list will track subject IDs for each trial (optional).

# Loop through all the cleaned .mat files in the output directory
for file in os.listdir(output_dir):  
    # Iterate over all files in the directory.
    if file.endswith('.mat'):  
        # Process only files with a .mat extension.

        # Construct the full file path
        file_path = os.path.join(output_dir, file)  
        # Combines the directory path and file name to create the full path.

        # Load the contents of the .mat file
        data = loadmat(file_path)  
        # Reads the .mat file into a Python dictionary-like structure.

        # Extract the cleaned signal data
        signal = data['cleaned_data']  
        # 'cleaned_data' is the key used to save the cleaned data in the previous script.
        # Ensure this matches the actual key used when saving the cleaned data.

        # Add the signal data to the list of all signals
        all_signals.append(signal)  
        # Append the signal (a 3D NumPy array) to the `all_signals` list.
        # Each element in `all_signals` represents the data from one file.

        # Optionally, track subject IDs for each trial
        subject_id = file.split('_')[0]  
        # Extract the subject ID from the file name. 
        # Assumes that the file name format is something like 'subjectID_trial.mat',
        # and the subject ID is the first part before the underscore.

        subject_ids.extend([subject_id] * signal.shape[2])  
        # Extend the `subject_ids` list by adding the subject ID for each trial in the file.
        # `signal.shape[2]` gives the number of trials in the signal.

```

```python
# Combine all the signal data into a single 3D NumPy array
all_signals = np.concatenate(all_signals, axis=2)  
# `np.concatenate` combines multiple arrays along a specified axis.
# Here, `axis=2` means the combination is done along the trials axis.
# This merges all the trials from all files into one large 3D array.

# Print the shape of the combined data
print(f"Combined data shape: {all_signals.shape}")  
# This outputs the shape of the combined data array in the format:
# (number of channels, number of time points, total number of trials).
```

```python
# Prepare the data for machine learning by reshaping it into 2D format

# Get the total number of trials (samples)
samples = all_signals.shape[2]  
# The third dimension of `all_signals` represents the number of trials (samples).

# Calculate the number of features
features = all_signals.shape[0] * all_signals.shape[1]  
# The first dimension is the number of channels, and the second is the number of time points.
# Multiplying them gives the total number of features for each trial.

# Reshape the data for machine learning
reshaped_data = all_signals.transpose(2, 0, 1).reshape(samples, features)  
# Step 1: `transpose(2, 0, 1)` rearranges the axes to (trials, channels, time points).
# This ensures that trials become the first dimension (rows).
# Step 2: `.reshape(samples, features)` flattens each trial (channels x time points) into a single vector.
# The resulting array is 2D with shape (samples, features).

# Print the shape of the reshaped data
print(f"Reshaped data shape: {reshaped_data.shape}")  
# The output shape is (samples, features), where:
# - `samples` is the total number of trials (rows).
# - `features` is the flattened data for each trial (columns).
```

```python
import os
from scipy.io import loadmat

# Load labels.csv
labels_df = pd.read_csv('C:/Users/DELL/mBTI_Data/labels.csv')

# Directory with cleaned .mat files
cleaned_dir = 'C:/Users/DELL/mBTI_Data/Data_Cleaned'

# Map SubID to ControlEquals1
labels_dict = labels_df.set_index('SubID')['ControlEquals1'].to_dict()

# Match trials to labels
matched_data = []
for file in os.listdir(cleaned_dir):
    if file.endswith('.mat'):
        # Extract SubID from file name (assumes filenames are like "cleaned_3001_1_Rest.mat")
        sub_id = int(file.split('_')[1])  # Extract SubID from the filename
        if sub_id in labels_dict:
            label = labels_dict[sub_id]
            matched_data.append({'File': file, 'Label': label})

# Convert matched data to a DataFrame
matched_df = pd.DataFrame(matched_data)
matched_df.to_csv('C:/Users/DELL/mBTI_Data/matched_trials.csv', index=False)
print("Matched trials and labels saved.")
```

```python
import numpy as np  # Library for numerical computations and handling arrays.
from sklearn.preprocessing import StandardScaler  # Used to normalize the data (standard scaling).

# Initialize empty lists to store data and labels
all_data = []  # Will store reshaped data from all files.
all_labels = []  # Will store corresponding labels for the data.

# Specify the directory containing the cleaned .mat files
cleaned_dir = 'C:/Users/DELL/mBTI_Data/Data_Cleaned'  
# Update this path to the folder where cleaned files are stored.

# Loop through all rows in the `matched_df` DataFrame
for _, row in matched_df.iterrows():  
    # `matched_df` is assumed to be a DataFrame containing file names and their corresponding labels.
    # `iterrows()` iterates over each row of the DataFrame.

    file_path = os.path.join(cleaned_dir, row['File'])  
    # Construct the full path to the cleaned .mat file for the current row.
    label = row['Label']  
    # Extract the label for the current file.

    # Load the cleaned data from the .mat file
    cleaned_data = loadmat(file_path)['cleaned_data']  
    # `loadmat` reads the .mat file, and `['cleaned_data']` extracts the cleaned data array.

    # Reshape the 3D cleaned data to 2D for machine learning
    samples = cleaned_data.shape[2]  
    # `samples` is the number of trials (third dimension of the array).
    features = cleaned_data.shape[0] * cleaned_data.shape[1]  
    # `features` is the total number of features (channels x time points).
    reshaped_data = cleaned_data.transpose(2, 0, 1).reshape(samples, features)  
    # Step 1: Transpose the data to reorder axes to (samples, channels, time points).
    # Step 2: Reshape each trial into a single feature vector.

    # Append the reshaped data and corresponding labels
    all_data.append(reshaped_data)  
    # Add the reshaped data from the current file to the `all_data` list.
    all_labels.extend([label] * samples)  
    # Extend the `all_labels` list with the label repeated for each trial.

# Combine all reshaped data into a single NumPy array
all_data = np.vstack(all_data)  
# `np.vstack` stacks the reshaped arrays from all files vertically into a single array.
# The resulting array has a shape of (total_samples, features).

# Convert the labels list into a NumPy array
all_labels = np.array(all_labels)  
# Convert the list of labels into a NumPy array for easier processing.

# Normalize the combined data
scaler = StandardScaler()  
# Initialize the standard scaler, which normalizes data to have a mean of 0 and a standard deviation of 1.
normalized_data = scaler.fit_transform(all_data)  
# Normalize the data using the scaler.

# Print the final shapes of the dataset and labels
print(f"Final dataset shape: {normalized_data.shape}")  
# Output the shape of the normalized dataset (samples, features).
print(f"Labels shape: {all_labels.shape}")  
# Output the shape of the labels array (samples,).
```

```python
from sklearn.model_selection import train_test_split  
# Import the `train_test_split` function from scikit-learn to split data into training and test sets.

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    normalized_data,  # The normalized feature data (samples x features).
    all_labels,       # The corresponding labels for the data.
    test_size=0.2,    # Specify the proportion of the dataset to include in the test set (20% here).
    random_state=42   # Random seed for reproducibility of the split. Any fixed number ensures the same split each time.
)

# Print the shapes of the training and test sets
print(f"Training set shape: {X_train.shape}, Test set shape: {X_test.shape}")  
# Output the shape of the training set (X_train) and test set (X_test).
# X_train and X_test contain the feature data for training and testing, respectively.
# y_train and y_test contain the corresponding labels.
```

```python
from sklearn.ensemble import RandomForestClassifier  # Import Random Forest Classifier
from sklearn.metrics import classification_report, accuracy_score  # Metrics to evaluate the model

# Initialize and train a Random Forest classifier
clf = RandomForestClassifier(random_state=42)  
# Create a Random Forest model with a fixed random state for reproducibility.
# By default, the classifier uses 100 trees. You can customize this with the `n_estimators` parameter.

clf.fit(X_train, y_train)  
# Train the Random Forest model using the training data (`X_train`) and their labels (`y_train`).

# Predict the labels for the test set
y_pred = clf.predict(X_test)  
# Use the trained model to predict labels for the test data (`X_test`).

# Calculate the accuracy of the model
print("Accuracy:", accuracy_score(y_test, y_pred))  
# `accuracy_score` computes the percentage of correctly predicted labels.

# Print a detailed classification report
print("Classification Report:")  
# The classification report provides precision, recall, F1-score, and support for each class.
print(classification_report(y_test, y_pred))  
# Generate the classification report, comparing true labels (`y_test`) with predicted labels (`y_pred`).
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, accuracy_score

# Evaluate accuracy on the training set
train_accuracy = clf.score(X_train, y_train)
print(f"Training Accuracy: {train_accuracy:.2f}")

# Evaluate accuracy on the test set
test_accuracy = clf.score(X_test, y_test)
print(f"Test Accuracy: {test_accuracy:.2f}")

# Check if the model might be overtrained
if train_accuracy - test_accuracy > 0.1:  # Threshold can be adjusted
    print("Warning: The model may be overfitting. Test accuracy is significantly lower than training accuracy.")

# Perform 5-fold cross-validation to evaluate overall model efficiency
cv_scores = cross_val_score(clf, np.vstack([X_train, X_test]), np.hstack([y_train, y_test]), cv=5)
print(f"Cross-Validation Scores: {cv_scores}")
print(f"Mean Cross-Validation Accuracy: {cv_scores.mean():.2f}")

# Get detailed classification report on the test set
y_pred = clf.predict(X_test)
print("\nClassification Report (Test Set):")
print(classification_report(y_test, y_pred))

# Overall Efficiency Evaluation
overall_accuracy = accuracy_score(y_test, y_pred)
print(f"\nOverall Model Accuracy: {overall_accuracy:.2f}")
```
